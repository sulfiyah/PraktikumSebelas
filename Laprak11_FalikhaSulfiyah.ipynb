```python
import math

print("1. Menghitung pangkat")
angka = int(input("Masukkan angka: "))
pangkat = int(input("Masukkan pangkat: "))
hasil = math.pow(angka, pangkat)
print(f"{angka} pangkat {pangkat} adalah: {hasil}")

print("2. Menghitung sinus dari sudut")
sudut = int(input("Masukkan sudut: "))
sinus = math.sin(sudut)
print(f"Sinus dari {sudut} adalah: {sinus}")

print("3. Menghitung cosinus dari sudut")
sudut = int(input("Masukkan sudut: "))
cosinus = math.cos(sudut)
print(f"Cosinus dari {sudut} adalah: {cosinus}")

print("4. Menghitung tangen dari sudut")
sudut = int(input("Masukkan sudut: "))
tangen = math.tan(sudut)
print(f"Tangen dari {sudut} adalah: {tangen}")

print("5. Menghitung akar kuadrat")
angka = int(input("Masukkan angka: "))
akar = math.sqrt(angka)
print(f"Akar kuadrat dari {angka} adalah: {akar}")
```
